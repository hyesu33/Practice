```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# Quiz questions
questions_data = [
    {
        "question": "Sarah is a 15-year-old girl who sneaks into the library during the day.",
        "answer": False,
        "explanation": "Sarah sneaks into the library at night, not during the day."
    },
    {
        "question": "The library in Willowby is said to be enchanted.",
        "answer": True,
        "explanation": "The story clearly describes the library as rumored to be enchanted."
    },
    {
        "question": "At midnight, the characters in the books come to life and talk.",
        "answer": True,
        "explanation": "At midnight, the books whisper and characters step out of the pages."
    },
    {
        "question": "Sarah met Harry Potter in the Midnight Library.",
        "answer": False,
        "explanation": "Harry Potter is not mentioned; she met characters like Alice and pirates from Treasure Island."
    },
    {
        "question": "The characters return to their pages as dawn approaches.",
        "answer": True,
        "explanation": "As dawn approaches, the characters return to their pages."
    }
]

# Shuffle questions
questions = random.sample(questions_data, len(questions_data))
user_answers = [None] * len(questions)
current_index = [0]

# Widgets
question_label = widgets.HTML()
radio_buttons = widgets.RadioButtons(options=["True", "False"], layout={'width': 'max-content'})
submit_button = widgets.Button(description="Submit Answer", button_style='info')
feedback_label = widgets.HTML()
navigation_buttons = widgets.HBox()
prev_button = widgets.Button(description="Previous")
next_button = widgets.Button(description="Next")
score_button = widgets.Button(description="Show Final Score", button_style='success')

def update_ui():
    i = current_index[0]
    question_label.value = f"<b>Question {i + 1} of {len(questions)}:</b><br>{questions[i]['question']}"
    if user_answers[i] is not None:
        radio_buttons.value = "True" if user_answers[i] else "False"
    else:
        radio_buttons.value = None
    feedback_label.value = ""

def on_submit_clicked(b):
    i = current_index[0]
    if radio_buttons.value is None:
        feedback_label.value = "<span style='color:red'>Please select an answer.</span>"
        return
    user_answer = True if radio_buttons.value == "True" else False
    user_answers[i] = user_answer
    correct_answer = questions[i]['answer']
    if user_answer == correct_answer:
        feedback_label.value = f"<span style='color:green'>Correct!</span><br><i>Explanation: {questions[i]['explanation']}</i>"
    else:
        feedback_label.value = f"<span style='color:red'>Incorrect.</span><br><i>Explanation: {questions[i]['explanation']}</i>"

def on_prev_clicked(b):
    if current_index[0] > 0:
        current_index[0] -= 1
        update_ui()

def on_next_clicked(b):
    if current_index[0] < len(questions) - 1:
        current_index[0] += 1
        update_ui()

def on_show_score(b):
    score = sum([1 for i, q in enumerate(questions) if user_answers[i] == q["answer"]])
    feedback_label.value = f"<h4>Your final score: {score} out of {len(questions)}</h4>"

# Bind button events
submit_button.on_click(on_submit_clicked)
prev_button.on_click(on_prev_clicked)
next_button.on_click(on_next_clicked)
score_button.on_click(on_show_score)

# Display everything
update_ui()
navigation_buttons.children = [prev_button, next_button, score_button]
display(question_label, radio_buttons, submit_button, feedback_label, navigation_buttons)
```
